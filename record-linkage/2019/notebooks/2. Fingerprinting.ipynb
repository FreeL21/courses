```python
import re
import csv
from collections import defaultdict
```

# Fonction de fingerprint

```python
SQUEEZE_REGEX = r'\s+'

def normalize(s):
    return re.sub(SQUEEZE_REGEX, ' ', s.lower().strip())
```

```python
# Test
normalize('Dupond'), normalize('DUPOND     ')
```

```python
def fingerprint(s):
    s = normalize(s).replace(',', '')
    s = set(s.split(' '))
    s = sorted(s)
    
    return ' '.join(s)
```

```python
fingerprint('University, of of of north CaRoLiNa')
```

# Clustering

```python
# Chargement du fichier
with open('../data/dirty-figshare.csv') as f:
    reader = csv.DictReader(f)
    journals = set(line['Journal title'] for line in reader)
```

```python
def clustering(strings):
    groups = defaultdict(list)
    
    for s in strings:
        groups[fingerprint(s)].append(s)
        
    for cluster in groups.values():
        if len(cluster) > 1:
            yield cluster
```

```python
# Test
for i, cluster in enumerate(clustering(journals)):
    print('Cluster n°%i' % (i + 1))
    
    for item in cluster:
        print('  • %s' % item)
    
    print()
```
